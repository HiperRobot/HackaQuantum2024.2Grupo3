```python
import numpy as np
import pandas as pd
```

```python
import os

caminho_dos_arquivos = './'
# Lista de arquivos CSV no diretório
arquivos_csv = [f for f in os.listdir(caminho_dos_arquivos) if f.endswith('.csv')]

# Criar uma lista vazia para armazenar os DataFrames
dados_fundos = []

# Iterar sobre os arquivos CSV e carregá-los
for arquivo in sorted(arquivos_csv):  # Ordenar por nome de arquivo para garantir a ordem correta dos dias
    caminho_completo = os.path.join(caminho_dos_arquivos, arquivo)
    
    # Carregar o CSV em um DataFrame
    df = pd.read_csv(caminho_completo)
    
    
    # Adicionar o DataFrame à lista
    dados_fundos.append(df)

# Concatenar todos os DataFrames em um único DataFrame
dados_concatenados = pd.concat(dados_fundos, ignore_index=True)

```
